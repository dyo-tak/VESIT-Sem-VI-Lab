```python
import pandas as pd
```

```python
deliveries = pd.read_csv('../data/raw/deliveries.csv')
deliveries.head()
```

```python
mi_csk = deliveries[(deliveries['batting_team'].isin(['Mumbai Indians', 'Chennai Super Kings'])) & (deliveries['bowling_team'].isin(['Mumbai Indians', 'Chennai Super King'])) & (deliveries['is_super_over'] == 0)]
```

```python
mi_csk = mi_csk.drop(['match_id', 'batsman', 'non_striker', 'bowler', 'batsman_runs','is_super_over', 'wide_runs', 'bye_runs', 'legbye_runs', 'noball_runs', 'penalty_runs', 'extra_runs', 'player_dismissed', 'dismissal_kind', 'fielder'], axis=1)
mi_csk.head(2)
```

```python
mi_csk = pd.get_dummies(mi_csk, columns=['batting_team'], prefix=['batting_team'])
```

```python
mi_csk = mi_csk.drop(['bowling_team'], axis=1)
mi_csk.rename(columns={'batting_team_Chennai Super Kings': 'mi_batting'}, inplace=True)
```

```python
mi_csk = mi_csk.replace({True: 0})
```

```python
mi_csk.head(2)
```

```python
mi_csk.to_csv('../data/cleaned/mi_csk.csv', index=False)
```
