```r
library(ggplot2)
library(dplyr)
library(broom)
```

```r
model <- lm(Y.house.price.of.unit.area ~ X1.transaction.date + X2.house.age +
              X3.distance.to.the.nearest.MRT.station + X4.number.of.convenience.stores +
              X5.latitude + X6.longitude, data = df)

summary(model)
```

```r
# Plot the regression model
plot(df$Y.house.price.of.unit.area, fitted(model),
     main="Regression Model", xlab="Observed Values", ylab="Fitted Values")

```

```r
ggplot(df, aes(x = Y.house.price.of.unit.area, y = fitted(model))) +
  geom_point() +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  ggtitle("Regression Model") +
  xlab("Observed Values") +
  ylab("Fitted Values") +
  theme_minimal()
```
