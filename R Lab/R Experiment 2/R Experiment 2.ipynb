```r
library('tidyverse')
library('ggplot2')
```

```r
bike_buyers <- read.csv('./bike_buyers_clean.csv', header=T, na.strings='')
head(bike_buyers)
```

## Experiment 2a

```r
hist(bike_buyers$Income, col="#FFD699")
```

## Experiment 2b

```r
education_counts <- table(bike_buyers$Education)

pie(education_counts, labels = names(education_counts),
    main = "Education Distribution")
```

```r
region_count <- table(bike_buyers$Region)

pie(region_count, labels = names(region_count),
    main = "Region Distribution")
```
